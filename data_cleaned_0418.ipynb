```python
#delete $ in price and then change price to float type
def format_price(input_price):
    if input_price[0]=='$':
        return input_price[1:]
    elif input_price =='0':
        return input_price        

google['Price']=google['Price'].apply(format_price).apply(float)
```

```python
#change all app sizes to MB and then change to float type
#change "varies with device" to NaN
def format_size(input_size):
    if input_size=='Varies with device':
        return np.NaN
    elif input_size[-1]=='M':
        return float(input_size[:-1])
    elif input_size[-1]=='k':
        return float(input_size[:-1])/1000

google['Size']=google['Size'].apply(format_size)
```

```python
def format_installs(input_installs):
    if input_installs[-1]=='+':
        input_installs=input_installs[:-1]
        if ',' in input_installs:
            return input_installs.replace(',','')
        else:
            return input_installs


google['Installs']=google['Installs'].apply(format_installs).apply(int)
```

```python
google['Reviews']=google['Reviews'].apply(int)
```

```python
#only 1 data entry of content rating is "unrated," so change it directly
#df['Content Rating'].unique()
#df[df['Content Rating']=='Unrated']
google.loc[8266,'Content Rating']='Teen'
```

```python
#convert to apple's rule: 4+, 9+, 12+, 17+
def convert_content(x):
    if x=="Everyone":
        return "4+"
    elif x=="Everyone 10+":
        return "9+"
    elif x=="Teen":
        return "12+"
    elif x=="Mature 17+" or "Adults only 18+":
        return "17+"

google['Content Rating']=google['Content Rating'].apply(convert_content)
```

```python
# To modify category names for apps in Apple appstore. Save the new category names in newCategory.
# Now the two dataframes shall have the same variables for newCategory.
apple['newCategory']=np.zeros(len(apple['prime_genre']))
for i in range(len(apple['prime_genre'])):
    if apple['prime_genre'].iloc[i] == 'Music':
        apple['newCategory'].iloc[i] = 'Entertainment'
    elif apple['prime_genre'].iloc[i] == 'Book':
        apple['newCategory'].iloc[i] = 'Books & Reference'
    elif apple['prime_genre'].iloc[i] == 'Reference':
        apple['newCategory'].iloc[i] = 'Books & Reference'
    elif apple['prime_genre'].iloc[i] == 'Catalogs':
        if 'Photo' in apple['track_name'].iloc[i]:
            apple['newCategory'].iloc[i] = 'Photo & Video'
        elif 'Movie' in apple['track_name'].iloc[i]:
            apple['newCategory'].iloc[i] = 'Entertainment'
        elif 'Classifieds' in apple['track_name'].iloc[i]:
            apple['newCategory'].iloc[i] = 'Shopping'
        elif 'Game' in apple['track_name'].iloc[i]:
            apple['newCategory'].iloc[i] = 'Games'
        elif 'Ringtone' in apple['track_name'].iloc[i]:
            apple['newCategory'].iloc[i] = 'Entertainment'
        else:
            apple.drop(apple.index[i],inplace=True)
    else:
        apple['newCategory'].iloc[i] = apple['prime_genre'].iloc[i]
```
