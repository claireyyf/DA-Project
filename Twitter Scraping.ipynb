```python
from bs4 import BeautifulSoup
import requests
```

```python
# #brexit since:2016-06-01 until:2019-04-05
url = 'https://twitter.com/search?l=&q=%23brexit%20since%3A2016-06-01%20until%3A2019-04-05&src=typd'
response = requests.get(url)
response.status_code
```

```python
content = BeautifulSoup(response.content,'lxml')
print(content)
```
